```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.ExcelFile(r'Data - FF.xlsx')
```

```python
price_sheet = pd.read_excel(df,'Price',skiprows=1,header = 0, 
                            index_col = 0, parse_dates = True)
df_price_fr = pd.read_excel(df,"Price", header = None, index_col = 0, 
                          parse_dates = True, nrows = 1)

bv_sheet = pd.read_excel(df,'Bookvalue',skiprows=1,header = 0, 
                            index_col = 0, parse_dates = True)
df_bv_fr = pd.read_excel(df,"Bookvalue", header = None, index_col = 0, 
                          parse_dates = True, nrows = 1)

mv_sheet = pd.read_excel(df,'Marketvalue',skiprows=1,header = 0, 
                            index_col = 0, parse_dates = True)
df_mv_fr = pd.read_excel(df,"Marketvalue", header = None, index_col = 0, 
                          parse_dates = True, nrows = 1)                          
```

```python
df_bv_fr
```

```python
df_price_fr.columns = price_sheet.columns
df_bv_fr.columns = bv_sheet.columns
df_mv_fr.columns = mv_sheet.columns
```

```python
# Price table
price = pd.concat([df_price_fr,price_sheet])
price.index.set_names('Time',inplace=True)
# Book value table
book_value = pd.concat([df_bv_fr,bv_sheet])
book_value.index.set_names('Time',inplace=True)
# Market value table
market_value = pd.concat([df_mv_fr,mv_sheet])
market_value.index.set_names('Time',inplace=True)
```

```python
# Drop columns have full of NaN data
price.dropna(how='all',axis=1, inplace=True)
book_value.dropna(how='all',axis=1, inplace=True)
market_value.dropna(how='all',axis=1, inplace=True)
```

```python
# Rename the columns
df_1, df_2, df_3 = pd.DataFrame({}), pd.DataFrame({}), pd.DataFrame({})

for i in price.columns:
    df_1[i.replace('(P)','').replace('VT:','')] = price[i]

for j in book_value.columns:
    df_2[j.replace('(WC05476)','').replace('VT:','')] = book_value[j]

for k in market_value.columns:
    df_3[k.replace('(MV)','').replace('VT:','')] = market_value[k]


price = df_1.copy()
book_value = df_2.copy()
market_value = df_3.copy()
```

```python
def drop_diff_stock(df):
    list_diff_stock = []
    for i in df.columns:
        if len(i) != 3:
            list_diff_stock.append(i)
    df.drop(list_diff_stock,axis=1, inplace=True)
    return df


price = drop_diff_stock(price)
book_value = drop_diff_stock(book_value)
market_value = drop_diff_stock(market_value)
```

```python
def check_fund_pofolio(df):
    check = 0
    for i in df.iloc[0]:
        if 'etf' in i:
            check += 1
    if check != 0:
        return False
    return True
    
print(check_fund_pofolio(price))
print(check_fund_pofolio(market_value))
print(check_fund_pofolio(book_value))
```

```python
def drop_delist_stock(df):
    list_dead_stock = []
    j = -1
    for i in df.iloc[0]:
        j += 1
        if 'DELIST' in i:
            list_dead_stock.append(df.iloc[0].index[j])
    df.drop(list_dead_stock,axis=1,inplace=True)
    return df

price = drop_delist_stock(price)
book_value = drop_delist_stock(book_value)
market_value = drop_delist_stock(market_value)
```

```python
# Delete first row
price = price.drop(price.index[0])
market_value = market_value.drop(market_value.index[0])
book_value = book_value.drop(book_value.index[0])

# Set index's type
price.index = price.index.astype('datetime64[ns]')
market_value.index = market_value.index.astype('datetime64[ns]')
book_value.index = book_value.index.astype('datetime64[ns]')
```

```python
price = price.resample('BM').last()
market_value = market_value.resample('BM').last()
```

```python
price.dropna(how='all',inplace=True)
book_value.dropna(how='all',inplace=True)
market_value.dropna(how='all',inplace=True)
```

```python
price = price.loc['2009-01-01':'2020-12-31']
book_value = book_value.loc['2009-01-01':'2020-12-31']
market_value = market_value.loc['2009-01-01':'2020-12-31']
```
